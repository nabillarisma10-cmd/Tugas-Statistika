# 1. Stratified Random Sampling berdasarkan genre, lalu mengambil 30 data teratas dari hasil sampling.

```
import pandas as pd

# Membaca dataset
Genre = pd.read_csv('/content/drive/MyDrive/MovieData-2018-2020.csv')

# Stratified Random Sampling by genre
jumlah_sampel = 30
jumlah_per_genre = (Genre['genre'].value_counts(normalize=True) * jumlah_sampel).round().astype(int)

stratified_sample = Genre.groupby('genre', group_keys=False).apply(
    lambda x: x.sample(n=min(len(x), jumlah_per_genre[x.name]), random_state=42)
).head(jumlah_sampel)

print("Stratified Random Sampling:\n", stratified_sample)
print("Jumlah data sample:", len(stratified_sample))
```

#2. Hitunglah Estimasi (Statistics - Estimating Population Means) Budget Film menggunakan data sample,

```
import scipy.stats as stats
import math
from sklearn.model_selection import train_test_split

# Mengambil sample 20% dari data
sample_data, _ = train_test_split(Genre, test_size=0.2, random_state=10)
budget_sample = sample_data['budget'].dropna()

# Specify sample mean (x_bar), sample standard deviation (s), sample size (n) and confidence level
x_bar = budget_sample.mean()
s = budget_sample.std()
n = len(budget_sample)
confidence_level = 0.95

# Calculate alpha, degrees of freedom (df), the critical t-value, and the margin of error
alpha = (1-confidence_level)
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf(1-alpha/2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value: {:.3f}".format(critical_t))
print("Margin of Error: {:.3f}".format(margin_of_error))
print("Confidence Interval: [{:.3f},{:.3f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.3f} and {:.3f}".format(lower_bound,upper_bound))
```
